```python
import numpy as np
from scipy.signal import savgol_filter
from utils import save_record, load_record
import matplotlib.pyplot as plt
```

```python
record_dir = "records/"
error_rates = np.linspace(0.02, 0.2, 10)
```

# Compare learning rates

## Bit-flip error

```python
x = range(1, 1001)
y1 = load_record(record_dir + "loss_0_3_001.csv")
y2 = load_record(record_dir + "loss_0_3_005.csv")
y3 = load_record(record_dir + "loss_0_3_010.csv")
y4 = load_record(record_dir + "loss_0_3_015.csv")
y5 = load_record(record_dir + "loss_0_3_020.csv")
y6 = load_record(record_dir + "loss_0_3_025.csv")

y1 = savgol_filter(y1, 25, 3, mode= 'nearest')
y2 = savgol_filter(y2, 25, 3, mode= 'nearest')
y3 = savgol_filter(y3, 25, 3, mode= 'nearest')
y4 = savgol_filter(y4, 25, 3, mode= 'nearest')
y5 = savgol_filter(y5, 25, 3, mode= 'nearest')
y6 = savgol_filter(y6, 25, 3, mode= 'nearest')

plt.plot(x, y1, label='lr = 0.01')
plt.plot(x, y2, label='lr = 0.05')
plt.plot(x, y3, label='lr = 0.1')
plt.plot(x, y4, label='lr = 0.15')
plt.plot(x, y5, label='lr = 0.2')
plt.plot(x, y6, label='lr = 0.25')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## Depolarizing error

```python
x = range(1, 1001)
y1 = load_record(record_dir + "loss_1_3_001.csv")
y2 = load_record(record_dir + "loss_1_3_005.csv")
y3 = load_record(record_dir + "loss_1_3_010.csv")
y4 = load_record(record_dir + "loss_1_3_015.csv")
y5 = load_record(record_dir + "loss_1_3_020.csv")
y6 = load_record(record_dir + "loss_1_3_025.csv")

y1 = savgol_filter(y1, 25, 3, mode= 'nearest')
y2 = savgol_filter(y2, 25, 3, mode= 'nearest')
y3 = savgol_filter(y3, 25, 3, mode= 'nearest')
y4 = savgol_filter(y4, 25, 3, mode= 'nearest')
y5 = savgol_filter(y5, 25, 3, mode= 'nearest')
y6 = savgol_filter(y6, 25, 3, mode= 'nearest')

plt.plot(x, y1, label='lr = 0.01')
plt.plot(x, y2, label='lr = 0.05')
plt.plot(x, y3, label='lr = 0.1')
plt.plot(x, y4, label='lr = 0.15')
plt.plot(x, y5, label='lr = 0.2')
plt.plot(x, y6, label='lr = 0.25')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# Bit-flip error

```python
x = error_rates
y1 = load_record(record_dir + "az_0_3.csv")
y2 = load_record(record_dir + "az_0_5.csv")
y3 = load_record(record_dir + "mcts_0_3.csv")
y4 = load_record(record_dir + "mcts_0_5.csv")

plt.xticks(error_rates)
plt.xlabel('Error Rate')
plt.xlim([0.01, 0.21])

plt.yticks([i * 0.1 for i in range(0, 11)])
plt.ylim([0, 1.05])
plt.ylabel('Win Rate')

plt.plot(x, y1, linestyle='-', color='C0', label='L=3 AlphaZero')
plt.plot(x, y2, linestyle='-', color='C1', label='L=5 AlphaZero')
plt.plot(x, y3, linestyle='--', color='C0', label='L=3 MCTS')
plt.plot(x, y4, linestyle='--', color='C1', label='L=5 MCTS')
plt.legend()
plt.show()
```
